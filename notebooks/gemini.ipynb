```python
from google import genai
from google.genai import types
import pathlib

# import .env variables
from dotenv import load_dotenv
import os
load_dotenv()
client = genai.Client()

# Retrieve and encode the PDF byte
filepath = pathlib.Path('data/section1.pdf')

# Upload the PDF using the File API
sample_file = client.files.upload(
  file=filepath,
)


prompt = "Convert this document to markdown format. Make sections to be the first level headers, subsections to be the second level headers, and so on. Exercises sections should be 3rd level headers. Theorems, examples, definitions, tables, exercises and other numbered items must contain start and end markers with their number, e.g., <<Theorem 2.1>> ... <</Theorem 2.1>>. Remove page headers and footers. Use latex syntax for all math expressions"
response = client.models.generate_content(
  model="gemini-2.5-flash",
  contents=[sample_file, prompt]
)


with open("section1.md", "w") as f:
    f.write(response.text)
```
